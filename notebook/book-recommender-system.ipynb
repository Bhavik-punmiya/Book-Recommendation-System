# Popularity based recommendation system    


```python
ratings
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
```

```python
ratings_with_name.isnull().sum()    
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popularity_df
```

```python
popular_df = popularity_df[popularity_df['num_ratings'] >= 250].sort_values('avg_rating', ascending=False)
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Year-Of-Publication', 'avg_rating', 'num_ratings', 'Image-URL-L', 'Image-URL-M']]
```

```python
popular_df
```

# Collaborative Filtering Based Recommendation System


```python
x  = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
```

```python
top_users = x[x].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(top_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values ='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```
